```python
import pandas as pd
import seaborn as sns

dados = pd.read_csv("../MICRODADOS_ENEM_2019_SAMPLE_43278.csv")
sns.set_palette('Set2')

dados.head()
```

```python
provas = ["NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_MT", "NU_NOTA_LC", "NU_NOTA_REDACAO"]

dados[provas].sum()
```

```python
dados[provas].sum(axis=1)

dados["NU_NOTA_TOTAL"] = dados[provas].sum(axis=1)
```

## Mostra a nota total dos participantes de acordo com a idade

```python
sns.scatterplot(data=dados, x="NU_NOTA_TOTAL", y="NU_IDADE")
```

## Mostra a quantidade de participantes em determinado estado civil

```python
sns.displot(data=dados, x="TP_ESTADO_CIVIL", bins=5, discrete=True)
```

## Faz um comparativo da idade dos participantes com seu estado civil

```python
sns.barplot(data=dados, x="TP_ESTADO_CIVIL", y="NU_IDADE", dodge=True)
```
